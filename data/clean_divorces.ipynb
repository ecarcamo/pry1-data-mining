## Limpieza del Dataset de divorcios

```python
import pandas as pd

df_div = pd.read_csv("files/Divorcios_Limpio.csv", sep=";", dtype=str)
```

FECHAS

```python
columnas_enteras = ["ANO", "MES", "DIA", "DEPTO", "MUNICIPIO"]

for col in columnas_enteras:
    df_div[col] = pd.to_numeric(df_div[col], errors="coerce").astype("Int64")
```

EDADES

```python
df_div["EDAD_HOM"] = pd.to_numeric(df_div["EDAD_HOM"], errors="coerce")
df_div["EDAD_MUJ"] = pd.to_numeric(df_div["EDAD_MUJ"], errors="coerce")

df_div.loc[df_div["EDAD_HOM"] == 999, "EDAD_HOM"] = pd.NA
df_div.loc[df_div["EDAD_MUJ"] == 999, "EDAD_MUJ"] = pd.NA

df_div = df_div[(df_div["EDAD_HOM"].between(10, 110)) | (df_div["EDAD_HOM"].isna())]
df_div = df_div[(df_div["EDAD_MUJ"].between(10, 110)) | (df_div["EDAD_MUJ"].isna())]

df_div["EDAD_HOM"] = df_div["EDAD_HOM"].astype("Int64")
df_div["EDAD_MUJ"] = df_div["EDAD_MUJ"].astype("Int64")
```

```python
print("Valores únicos EDAD_HOM:")
print(sorted(df_div["EDAD_HOM"].dropna().unique()))

print("Valores únicos EDAD_MUJ:")
print(sorted(df_div["EDAD_MUJ"].dropna().unique()))

print(df_div["EDAD_HOM"].dtype)
print(df_div["EDAD_MUJ"].dtype)
```

MUNICIPIOS

```python
df_div["MUNICIPIO"] = pd.to_numeric(df_div["MUNICIPIO"], errors="coerce")
df_div["MUNICIPIO"] = df_div["MUNICIPIO"].astype("category")

municipios = pd.read_csv("files/municipios.csv")
mapa_municipios = dict(
    zip(municipios["MUNICIPIO_CODIGO"], municipios["MUNICIPIO_NOMBRE"])
)

df_div["MUNICIPIO_NOMBRE"] = df_div["MUNICIPIO"].map(mapa_municipios)
df_div["MUNICIPIO_NOMBRE"] = df_div["MUNICIPIO_NOMBRE"].astype("category")
```

DEPARTAMENTOS

```python
df_div["DEPTO"] = pd.to_numeric(df_div["DEPTO"], errors="coerce")
df_div["DEPTO"] = df_div["DEPTO"].astype("category")

mapa_deptos = {
    1: "Guatemala",
    2: "El Progreso",
    3: "Sacatepéquez",
    4: "Chimaltenango",
    5: "Escuintla",
    6: "Santa Rosa",
    7: "Sololá",
    8: "Totonicapán",
    9: "Quetzaltenango",
    10: "Suchitepéquez",
    11: "Retalhuleu",
    12: "San Marcos",
    13: "Huehuetenango",
    14: "Quiché",
    15: "Baja Verapaz",
    16: "Alta Verapaz",
    17: "Petén",
    18: "Izabal",
    19: "Zacapa",
    20: "Chiquimula",
    21: "Jalapa",
    22: "Jutiapa"
}

df_div["DEPTO_NOMBRE"] = df_div["DEPTO"].map(mapa_deptos)
df_div["DEPTO_NOMBRE"] = df_div["DEPTO_NOMBRE"].astype("category")
```

ETNIA

```python
df_div["ETNIA_HOM"] = pd.to_numeric(df_div["ETNIA_HOM"], errors="coerce")
df_div["ETNIA_HOM"] = df_div["ETNIA_HOM"].astype("category")

df_div["ETNIA_MUJ"] = pd.to_numeric(df_div["ETNIA_MUJ"], errors="coerce")
df_div["ETNIA_MUJ"] = df_div["ETNIA_MUJ"].astype("category")

etnias = pd.read_csv("files/etnias.csv")
mapa_etnias = dict(
    zip(etnias["ETNIA_CODIGO"], etnias["ETNIA_NOMBRE"])
)

df_div["ETNIA_HOMBRE_NOMBRE"] = df_div["ETNIA_HOM"].map(mapa_etnias)
df_div["ETNIA_HOMBRE_NOMBRE"] = df_div["ETNIA_HOMBRE_NOMBRE"].astype("category")

df_div["ETNIA_MUJER_NOMBRE"] = df_div["ETNIA_MUJ"].map(mapa_etnias)
df_div["ETNIA_MUJER_NOMBRE"] = df_div["ETNIA_MUJER_NOMBRE"].astype("category")

df_div["ETNIA_HOM"] = df_div["ETNIA_HOM"].astype("Int64")
df_div["ETNIA_MUJ"] = df_div["ETNIA_MUJ"].astype("Int64")
```

```python
print(df_div["ETNIA_HOMBRE_NOMBRE"].value_counts(dropna=False))
print(df_div["ETNIA_MUJER_NOMBRE"].value_counts(dropna=False))

print("Códigos HOM sin mapear:")
print(df_div.loc[df_div["ETNIA_HOMBRE_NOMBRE"].isna(), "ETNIA_HOM"].unique())
```

ESCOLARIDAD

```python
df_div["ESCOLARIDAD_HOM"] = pd.to_numeric(df_div["ESCOLARIDAD_HOM"], errors="coerce")
df_div["ESCOLARIDAD_HOM"] = df_div["ESCOLARIDAD_HOM"].astype("category")

df_div["ESCOLARIDAD_MUJ"] = pd.to_numeric(df_div["ESCOLARIDAD_MUJ"], errors="coerce")
df_div["ESCOLARIDAD_MUJ"] = df_div["ESCOLARIDAD_MUJ"].astype("category")

escolaridad = pd.read_csv("files/nivel.csv")
mapa_escolaridad = dict(
    zip(escolaridad["NIVEL_CODIGO"], escolaridad["NIVEL_NOMBRE"])
)

df_div["ESCOLARIDAD_HOMBRE_NOMBRE"] = df_div["ESCOLARIDAD_HOM"].map(mapa_escolaridad)
df_div["ESCOLARIDAD_HOMBRE_NOMBRE"] = df_div["ESCOLARIDAD_HOMBRE_NOMBRE"].astype("category")

df_div["ESCOLARIDAD_MUJER_NOMBRE"] = df_div["ESCOLARIDAD_MUJ"].map(mapa_escolaridad)
df_div["ESCOLARIDAD_MUJER_NOMBRE"] = df_div["ESCOLARIDAD_MUJER_NOMBRE"].astype("category")

df_div["ESCOLARIDAD_HOM"] = df_div["ESCOLARIDAD_HOM"].astype("Int64")
df_div["ESCOLARIDAD_MUJ"] = df_div["ESCOLARIDAD_MUJ"].astype("Int64")
```

```python
print(df_div["ESCOLARIDAD_HOMBRE_NOMBRE"].value_counts(dropna=False))
print(df_div["ESCOLARIDAD_MUJER_NOMBRE"].value_counts(dropna=False))

print("Códigos HOM sin mapear:")
print(df_div.loc[df_div["ESCOLARIDAD_HOMBRE_NOMBRE"].isna(), "ESCOLARIDAD_HOM"].unique())
```

```python
df_div.to_csv("divorcios_limpio1.csv", index=False, encoding="utf-8")
```
