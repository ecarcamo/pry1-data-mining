```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sb
import sklearn.cluster as cluster
import sklearn.metrics as metrics
import sklearn.preprocessing
import scipy.cluster.hierarchy as sch
import skfuzzy as fuzz
import pylab
import sklearn.mixture as mixture
import pyclustertend 
import random
from scipy import stats
```

```python
df = pd.read_csv("files/Dataset_Unificado_Proyecto.csv", sep=";")
```

```python
#variables geográficas

df["DEPTO"] = pd.to_numeric(df["DEPTO"], errors="coerce")
df["MUNICIPIO"] = pd.to_numeric(df["MUNICIPIO"], errors="coerce")

df["DEPTO"] = df["DEPTO"].astype("category")
df["MUNICIPIO"] = df["MUNICIPIO"].astype("category")

mapa_deptos = {
    1: "Guatemala",
    2: "El Progreso",
    3: "Sacatepéquez",
    4: "Chimaltenango",
    5: "Escuintla",
    6: "Santa Rosa",
    7: "Sololá",
    8: "Totonicapán",
    9: "Quetzaltenango",
    10: "Suchitepéquez",
    11: "Retalhuleu",
    12: "San Marcos",
    13: "Huehuetenango",
    14: "Quiché",
    15: "Baja Verapaz",
    16: "Alta Verapaz",
    17: "Petén",
    18: "Izabal",
    19: "Zacapa",
    20: "Chiquimula",
    21: "Jalapa",
    22: "Jutiapa"
}

df["DEPTO_NOMBRE"] = df["DEPTO"].map(mapa_deptos)
df["DEPTO_NOMBRE"] = df["DEPTO_NOMBRE"].astype("category")

municipios = pd.read_csv("files/municipios.csv")
mapa_municipios = dict(
    zip(municipios["MUNICIPIO_CODIGO"], municipios["MUNICIPIO_NOMBRE"])
)

df["MUNICIPIO_NOMBRE"] = df["MUNICIPIO"].map(mapa_municipios)
df["MUNICIPIO_NOMBRE"] = df["MUNICIPIO_NOMBRE"].astype("category")

```

```python
#fecha

mapa_meses = {
    "Enero": 1,
    "Febrero": 2,
    "Marzo": 3,
    "Abril": 4,
    "Mayo": 5,
    "Junio": 6,
    "Julio": 7,
    "Agosto": 8,
    "Septiembre": 9,
    "Octubre": 10,
    "Noviembre": 11,
    "Diciembre": 12
}

df["MES"] = df["MES"].map(mapa_meses)

df["ANO"] = pd.to_numeric(df["ANO"], errors="coerce")
df["MES"] = pd.to_numeric(df["MES"], errors="coerce")
df["DIA"] = pd.to_numeric(df["DIA"], errors="coerce")
```

```python
#victima
df["VIC_SEXO"] = df["VIC_SEXO"].astype("category")
df["VIC_EDAD"] = pd.to_numeric(df["VIC_EDAD"], errors="coerce")
df["VIC_ESCOLARIDAD"] = df["VIC_ESCOLARIDAD"].astype("category")
df["VIC_EST_CIV"] = df["VIC_EST_CIV"].astype("category")
df["VIC_REL_AGR"] = df["VIC_REL_AGR"].astype("category")

#agresor
df["AGR_SEXO"] = df["AGR_SEXO"].astype("category")
df["AGR_EDAD"] = pd.to_numeric(df["AGR_EDAD"], errors="coerce")
df["AGR_GRUPET"] = df["AGR_GRUPET"].astype("category")

#de hecho
df["TIPAGRE"] = df["TIPAGRE"].astype("category")
df["MUNICIPIO"] = df["MUNICIPIO"].astype("category")
df["RECUR_DENUN"] = df["RECUR_DENUN"].astype("category")

#familiar
df["TOTAL_HIJOS"] = pd.to_numeric(df["TOTAL_HIJOS"], errors="coerce")

df.info()
```

```python
#edad
def limpiar_edad(col, min_edad=0, max_edad=120):
    col = pd.to_numeric(col, errors="coerce")

    col = col.where(
        (col >= min_edad) & (col <= max_edad),
        np.nan
    )

    return col

df["VIC_EDAD"] = limpiar_edad(df["VIC_EDAD"])
df["AGR_EDAD"] = limpiar_edad(df["AGR_EDAD"])

df["VIC_EDAD"].describe()
df["AGR_EDAD"].describe()

df["VIC_EDAD"].isna().sum()
df["AGR_EDAD"].isna().sum()
```
